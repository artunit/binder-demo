```python
# use requests for https call
import requests
from requests.auth import HTTPBasicAuth
```

In this case, basic apache authentication is used for accessing reverse proxy URL (the proxy allows the IP to be identified with the organization). Replace _USERID_ and _PW_ with the appropriate values. This would always have to be done on the fly since credentials could never be stored with a notebook.

```python
USERID='USERID HERE'
PW='PASSWORD HERE'
URL='https://OCUL_PROXY_HERE.ca/data/json/21615330/v2019inone/1_otuontttoocr.xml'
```

The data can now be accessed in the notebook.

```python
resp = requests.get(URL, auth=HTTPBasicAuth(USERID, PW))
json_vals = resp.json()
print('body: ',json_vals['article']['body'])
```
